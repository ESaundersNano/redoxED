```python
from sympy import symbols, exp, integrate, oo, simplify

u, x, x1, x2 = symbols('u x x1 x2', real=True)
phi1 = exp(-u**2 * (x - x1)**2)
phi2 = exp(-u**2 * (x - x2)**2)
integral = integrate(phi1 * phi2, (x, -oo, oo))
print(simplify(integral))
```
